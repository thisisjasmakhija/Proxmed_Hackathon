```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from skimage import measure
import SimpleITK as sitk

# A path to a T1-weighted brain .nii image:
t1_fn = r'C:\Users\DELL\Downloads\HYPODENSITY-DATA\HYPODENSITY-DATA\ProxmedImg021\ProxmedImg021_ROI.nii'

# Read the .nii image containing the volume with SimpleITK:
sitk_t1 = sitk.ReadImage(t1_fn)

# Access the numpy array:
t1 = sitk.GetArrayFromImage(sitk_t1)

# Function to display 3D volume
def plot_3d_volume(volume):
    threshold = (volume.min() + volume.max()) / 2  # Set threshold to the middle of the data range
    verts, faces, _, _ = measure.marching_cubes(volume, level=threshold)
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    mesh = Poly3DCollection(verts[faces], alpha=0.1, facecolors='cyan', edgecolor='r')
    ax.add_collection3d(mesh)
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")
    plt.show()

# Display the 3D volume
plot_3d_volume(t1)
```

```python
fig, ax = plt.subplots(10, 10,constrained_layout=True)
fig.canvas.set_window_title('4D Nifti Image')
fig.suptitle('4D_Nifti 10 slices 30 time Frames', fontsize=16)
#-------------------------------------------------------------------------------
mng = plt.get_current_fig_manager()
mng.full_screen_toggle()

for slice in range(10):
    # if your data in 4D, otherwise remove this loop
    for frame in range(10):
        ax[frame, slice].imshow(t1[:,:,slice,frame],cmap='gray', interpolation=None)
        ax[frame, slice].set_title("layer {} / frame {}".format(slice, frame))
        ax[frame, slice].axis('off')

plt.show() 
```
